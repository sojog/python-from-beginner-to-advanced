```python
teams = ["Real Madrid", "AC Milan", "Bayern Munchen"]
countries  = ['Spain', 'Italy', 'Germany']
trophies = [15, 7, 6]
big_teams_df = pd.DataFrame({
    'team':teams, 'country':countries, 'trophies':trophies
})
big_teams_df 
```

```python
df.drop(columns=['id'], inplace=True)
```

```python
the_new_df = pd.concat([df, big_teams_df])
```

```python
the_new_df = the_new_df.reset_index(drop=True)
```

```python
final_df =the_new_df.sort_values(['trophies', 'team'], ascending=(False, True)).reset_index(drop=True)
final_df
```
