```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]
plt.plot(x, y)
```

```python
x = [1, 2, 3, 4]
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.xlabel('This is X Axis')
plt.ylabel('This is Y Axis')
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.xlabel('This is X Axis')
plt.ylabel('This is Y Axis')
plt.plot(x, y, 'green')
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.xlabel('This is X Axis')
plt.ylabel('This is Y Axis')
plt.plot(x, y, "fuchsia") # https://matplotlib.org/stable/gallery/color/named_colors.html
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.xlabel('This is X Axis')
plt.ylabel('This is Y Axis')
plt.plot(x, y, "y") # https://matplotlib.org/stable/gallery/color/named_colors.html
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
```

```python
x = [1, 2, 3, 4]
y = [1, 10, 7, 1]
plt.scatter(x, y)
```

```python
x = [1, 2, 3, 4]
y = ["A", "B", "C", "D"]
plt.bar(x, y)
```
