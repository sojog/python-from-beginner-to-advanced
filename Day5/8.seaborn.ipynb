```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = [10,8]
tips_data = sns.load_dataset("tips")
tips_data.head()
```

Script 1:

```python
plt.rcParams["figure.figsize"] = [10,8]
sns.displot(tips_data["total_bill"])
```

Script 2:

```python
sns.displot(tips_data["tip"])
```

Script 3:

```python
sns.displot(tips_data["tip"], kde = False)
```

Script 4:

```python
sns.jointplot(x="total_bill", y="tip", data=tips_data)
```

Script 5:

```python
sns.jointplot(x="size", y="total_bill", data=tips_data,
kind = "reg")
```

Script 6:

```python
sns.pairplot(data=tips_data)
```

Script 7:

```python
sns.pairplot(data=tips_data, hue = "time")
```

Script 8:

```python
sns.rugplot(tips_data["total_bill"])
```
